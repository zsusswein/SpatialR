### This code takes cleaned mobility data in data/clean_mobility/counties which is organized by fips and turns them into weekly slices

```python
import pandas as pd
import glob
from os.path import exists
import datetime as dt
```

```python
# OPTION 1 (requires 1.5+ gigs of memory available)

# read in all by-county files and combine into a dataframe
path = 'data/clean_mobility/counties/'
files = glob.glob(path+'/*.csv')
files.sort()
df = pd.DataFrame()
for file in files:
    print(file)
    df_temp = pd.read_csv(file)
    
    df = pd.concat([df,df_temp], ignore_index=True)
df['week'] = pd.to_datetime(df.week, utc=True).dt.date

# output combined dataframe by weeks
weeks = df.week.unique()
weeks.sort()
for week in weeks:
    df_week = df[df.week == week]
    
    df_week.to_csv('data/clean_mobility/weeks_Shweta/'+str(week)+'.csv', index=False)
```

```python
# OPTION 2 (low memory but slower)

# read in each county file and output data to each week file for that county
path = 'data/clean_mobility/counties/'
files = glob.glob(path+'/*.csv')
files.sort()
df = pd.DataFrame()
for file in files:
    df = pd.read_csv(file)  
    df['week'] = pd.to_datetime(df.week, utc=True).dt.date
   
    weeks = df.week.unique()
    for week in weeks:
        df_week = df[df.week == week]
        if exists('weeks_Shweta2/'+str(week)+'.csv'):
            df_week.to_csv('data/clean_mobility/weeks_Shweta/'+str(week)+'.csv', mode='a', index=False, header=False)
        else:
            df_week.to_csv('data/clean_mobility/weeks_Shweta/'+str(week)+'.csv', index=False)
```
